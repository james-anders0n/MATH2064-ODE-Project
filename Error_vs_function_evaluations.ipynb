## Error vs Function Evaluations (different step-sizes for each method)

```julia
evals_per_step = Dict(
    "Euler" => 1,
    "RK2"   => 2,
    "RK4"   => 4
)

solver_methods = Dict(
    "Euler" => euler,
    "RK2"   => rk2,
    "RK4"   => rk4
)

fevals_common = [8, 16, 32, 64, 128, 256]

fevals = Dict{String, Vector{Float64}}()
errors_feval = Dict{String, Vector{Float64}}()

for (name, method) in solver_methods
    xvals = Float64[]
    yvals = Float64[]

    s = evals_per_step[name]

    for fe in fevals_common
        if fe % s == 0
            N = fe ÷ s
            h = (T - t0) / N

            t, y_num = method(f, t0, y0, T, h)

            push!(xvals, Float64(fe))
            push!(yvals, max_error(t, y_num, y_exact))
        end
    end

    fevals[name] = xvals
    errors_feval[name] = yvals
end

plot(
    fevals["Euler"], errors_feval["Euler"],
    xscale = :log10, yscale = :log10,
    marker = :circle, linewidth = 2, label = "Euler"
)
plot!(
    fevals["RK2"], errors_feval["RK2"],
    marker = :circle, linewidth = 2, label = "RK2"
)
plot!(
    fevals["RK4"], errors_feval["RK4"],
    marker = :circle, linewidth = 2, label = "RK4"
)

xlabel!("Function evaluations")
ylabel!("Maximum global error")
title!("Error vs function evaluations")

savefig("error_vs_function_evaluations2.png")
println("Saved plot to error_vs_function_evaluations2.png")
```
